## Naive Bayes
The Bayes' rule is called "Naive Bayes" if we assume each evidence makes an independent and equal contribution to the belief.

With the independent and equal contribution to the belief, Bayes' rule can transformed to the following:
<br/><br/>
$P(B_i|E) = \frac{P(B_i)P(e_1|B_i)P(e_2|B_i)P(e_3|B_i)\ldots P(e_d|B_i)}{\sum_{j=1}^n P(e_1|B_j)P(e_2|B_j)P(e_3|B_j)\ldots P(e_d|B_j)P(B_j)}$
<br/><br/>
where $E = \{ e_1, e_2, e_3, \ldots, e_d\}$ is a set of multiple pieces of evidence, $B_i$ is belief/class.
<br/><br/>
Suppose we further assume that we are only interested in knowing which brief has the highest probability value rather than the actual probability value. In that case, we can further simplify the equation by removing the denominator as follows:
<br/><br/>
$B_{NB} = argmax_{B_i}P(B_i)P(e_1|B_i)P(e_2|B_i)P(e_3|B_i)\ldots P(e_d|B_i)$
<br/><br/>
It is because the denominator remains constant for a given input.
<br/><br/>


## Underflow Prevention

Multiplying lots of probabilities between 0 and 1 by definition, can result in floating-point underflow.

log(xy) = log(x) + log(y), it is better to perform all computations by summing logs of probabilities rather than multiplying probabilities.

Brief with the highest final log probability score is still the most probable.

<br/>
$B_{NB} = argmax_{B_i} \left( logP(B_i) + \sum_{n=1}^d logP(e_n|B_i) \right)$

# Conclusion

The assumption made by Naive Bayes are <font color="red">NOT</font> generally correct in real-world situations. In fact, the independence assumption is never correct but often works well in practice.

## Minimax in general

* A **recursive** algorithm for choosing an optimal move for a player (maximizer), assuming the other (minimizer) is also playing optimally
* The maximizer tries to maximize the score while the minimizer tries to minimize it


## Steps of the Minimax algorithm
1. Apply utility function to each terminal state (i.e., assign numeric values), where a relatively large value (e.g., +1) is assigned if the maximizer wins, a small value (e.g., -1) is assigned if the minimizer wins, and an intermedium value (e.g., 0) is assigned if draw.
2. Given a game state, to decide the optimal move for the maximizer, call the Minimax algorithm to obtain the value for each possible move, which is the root of the tree.
3. Inside the Minimax algorithm, return the value if it reaches a terminal state; Otherwise, the maximizer/minimizer returns the max/min value among all possible moves by calling the Minimax algorithm recursively and alternately (the current player switches between the maximizer and minimizer). 
4. Finally, the root (i.e., the maximizer) obtains values of all possible moves, and it chooses the move with the largest value.

## Disadvantages of Minimax
1. Slow to reach goal state due to huge branching factor (traverse every node)
2. Search and evaluation of unnecessary nodes or branches
3. Both min and max players have lots of choices to decide from

## Alpha-Beta Pruning
* Cutting off the branches that are not worth exploring for a faster search, which addresses the second disadvantage of Minimax listed above

## Steps of Alpha-Beta Pruning
1. Define two values, $\alpha$ and $\beta$, to keep track of current upper and lower bounds for the maximizer and minimizer, respectively.
2. Pass current values of $\alpha$ and $\beta$ from the parent down to child nodes during the search.
3. Update values of $\alpha$ and $\beta$ with the current node value $V$, where the maximizer only updates $\alpha$ using $max(\alpha, V)$, and the minimizer only updates $\beta$ using $min(\beta, V)$.
4. If $\alpha > \beta$ satisfies, prune the remaining branches at the node.

## Alpha-Beta Pruning with Depth
Even with alpha-beta pruning, you may notice that the AI is not so "smart" when it knows it will lose for sure. However, we would like AI to act more like humans, i.e., it will try to delay the loss. Similarly, when AI can win the game immediately, it should do it rather than delaying the winning.

To achieve this, we can do a simple modification of alpha-beta pruning by further considering the depth of the game tree. Specifically, among the moves that lead to loss, it chooses the deepest branch; For the moves that lead to winning, it chooses the shallowest one.

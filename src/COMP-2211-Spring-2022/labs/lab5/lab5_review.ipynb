## __Response and Activation Function__

In a simple perceptron, it aggregates all inputs ($x_1, x_2, \ldots, x_n$) with the corresponding weights ($w_1, w_2, \ldots, w_n$) and apply bias ($\theta$).
>$Y(x_{1}, x_{2}, ...,x_{n}) = \sum_{i=1}^n(w_{i} \times x_{i}) + \theta$.

In this function, the $w_1, w_2, \ldots, w_n$, and $\theta$ are all trainable parameters, which will be updated through a training process.

After getting the value Y, we need to use an activation function, $f$, to judge the response. And the type of activation function can be various, here we may use 
>$response = f(Y) = 1 $  for  $Y>0$<br/>
>$response = f(Y) = 0 $  for  $Y\leq0$

for easy learning.



## __Weight update__

After getting the idea about getting result for inputs using perceptron, we need to know why perceptron learnable.

As we just mentioned, there are trainable parameters $w_1, w_2, \ldots, w_n$ and $\theta$ in the function. So what we will do is updating those initialized parameters when we go through the data samples we have.

For every data sample:
>Compute $f(Y(x_{1}, x_{2}, ...,x_{n}))$ <br/>
>Check if $f(Y)==T$, where $T$ is the true label for the data sample.<br/>
>If yes, $w_1, w_2, \ldots, w_n$ and $\theta$ stay unchanged.<br/>
>If not, $w_1, w_2, \ldots, w_n$ and $\theta$ will be updated according to following rules such that output of the perceptron for the new weights is closer to true label.<br/>
>>1. $w_1, w_2, \ldots, w_n$ will be updated according to value of input $x_1, x_2, \ldots, x_n$ and output difference $T-f(Y)$.
>>2. $\theta$ will be updated according to value of the output difference $T-f(Y)$. 

The number that controls how much the weights change is called the $learning$ $rate$.

The number of going through all training data samples is always called $epochs$.

Chosen $learning$ $rate$ and $max$ $epochs$ are important as they influence final performance and training time.





## __Stopping rule__

Now we get the idea about how to start the training process.

But, when shall we stop? How can I know the perceptron is ready?

Here are several rules we can follow:


*   Stop when meeting $max$ $training$ $time$. We can set a time limit for the training process.
*   Stop when meeting $max$ $epochs$. We can set the maximum number of rounds that the training process can run.
*   Stop when the $error$ is low enough. We can check the average error after each epoch. If the error is lower than a certain number, stop the training process.
*   Stop when $accuracy$ is high enough. Similar to the above one, we can count the accuracy level after each epoch. If the accuracy is higher than a certain number, stop the training process.
*   Stop when $weights$ do not update or update little. We can check the change of weights in each epoch, if the change is little, stop the training process.

The idea of stop rules is not to waste much resource and time on little improvement.

## __Limitation__

As a simple binary classification model, perceptron has its own drawback and problems.
 
>Perceptron is not suitable for the non-linear problem. 

Perceptron is like a straight line in the multi-dimension space to separate the data samples by their label. 

If the data samples can not be separated by the line, then a simple perceptron can not help classify it.

>Easy to be influenced by data distribution.

As a binary classification, simple perceptron suffers from data imbalance and insufficient data. 

Those situations will let perceptron wrongly learn the rule about classification, and it becomes easy to produce results like __all 0s__ or __all 1s__.
